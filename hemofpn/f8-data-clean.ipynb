```python
from PIL import Image
import numpy as np

# Load the RGB image
image_path = '/Users/jamesemilian/triage/hemobox/data/f8-data/masks_rgb/adt2_image_104.png'
image = Image.open(image_path)

# Convert the image to a numpy array (HxWx3 for RGB)
image_array = np.array(image)

# Define the target colors in RGB
color_1 = np.array([0x3c, 0x17, 0xff])  # #3c17ff
color_2 = np.array([0xff, 0xc1, 0x17])  # #ffc117

# Create a binary mask
binary_mask = np.zeros(image_array.shape[:2], dtype=np.uint8)

# Only consider the first 3 channels (RGB) for comparison
rgb_image_array = image_array[:, :, :3]
# Find the pixels that match either color_1 or color_2
binary_mask[np.all(rgb_image_array == color_1, axis=-1)] = 255
binary_mask[np.all(rgb_image_array == color_2, axis=-1)] = 255

# Convert the mask to have 4 channels (HxWx4)
binary_mask_rgba = np.stack([binary_mask]*4, axis=-1)

# Convert back to an image
binary_mask_image = Image.fromarray(binary_mask_rgba)

# Save the binary mask image
binary_mask_image.save('/Users/jamesemilian/triage/hemobox/data/f8-data/masks_gray/binary_bloodmask2.png')

print("Binary mask created and saved successfully!")
```

```python
import os
from PIL import Image
import numpy as np

# Define the source and destination folders
source_folder = '/Users/jamesemilian/triage/hemobox/data/f8-data/masks_rgb'
destination_folder = '/Users/jamesemilian/triage/hemobox/data/f8-data/masks_gray'

# Define the target colors in RGB
color_1 = np.array([0x3c, 0x17, 0xff])  # #3c17ff
color_2 = np.array([0xff, 0xc1, 0x17])  # #ffc117

# Ensure the destination folder exists
os.makedirs(destination_folder, exist_ok=True)

# Loop through all files in the source folder
for filename in os.listdir(source_folder):
    if filename.endswith(".png"):  # Only process .png files
        # Construct full file path
        image_path = os.path.join(source_folder, filename)
        
        # Load the RGB image
        image = Image.open(image_path)

        # Convert the image to a numpy array (HxWx3 for RGB)
        image_array = np.array(image)

        # Create a binary mask
        binary_mask = np.zeros(image_array.shape[:2], dtype=np.uint8)

        # Only consider the first 3 channels (RGB) for comparison
        rgb_image_array = image_array[:, :, :3]

        # Find the pixels that match either color_1 or color_2
        binary_mask[np.all(rgb_image_array == color_1, axis=-1)] = 255
        binary_mask[np.all(rgb_image_array == color_2, axis=-1)] = 255

        # Convert the mask to have 4 channels (HxWx4)
        binary_mask_rgba = np.stack([binary_mask] * 4, axis=-1)

        # Convert back to an image
        binary_mask_image = Image.fromarray(binary_mask_rgba)

        # Save the binary mask image in the destination folder
        save_path = os.path.join(destination_folder, filename)
        binary_mask_image.save(save_path)

        print(f"Binary mask for {filename} created and saved successfully!")

print("All images processed successfully!")
```

```python
import os
from PIL import Image
import numpy as np

# Define the source and destination folders
source_folder = '/Users/jamesemilian/triage/hemobox/data/f8-data-v2/2663250/masks_rgb'
destination_folder = '/Users/jamesemilian/triage/hemobox/data/f8-data-v2/2663250/masks_gray'

# Define the target colors in RGB
color_1 = np.array([0x3c, 0x17, 0xff])  # #3c17ff - blood_on_skin
color_2 = np.array([0xff, 0xc1, 0x17])  # #ffc117 - blood_on_ground
color_3 = np.array([0x35, 0xff, 0x1f])  # #35ff1f - blood_mask
color_4 = np.array([0xe3, 0x17, 0xff])  # #e317ff - pooled_blood


# Ensure the destination folder exists
os.makedirs(destination_folder, exist_ok=True)

# Loop through all files in the source folder
for filename in os.listdir(source_folder):
    if filename.endswith(".png"):  # Only process .png files
        # Construct full file path
        image_path = os.path.join(source_folder, filename)
        
        # Load the RGB image
        image = Image.open(image_path)

        # Convert the image to a numpy array (HxWx3 for RGB)
        image_array = np.array(image)

        # Create a binary mask
        binary_mask = np.zeros(image_array.shape[:2], dtype=np.uint8)

        # Only consider the first 3 channels (RGB) for comparison
        rgb_image_array = image_array[:, :, :3]

        # Find the pixels that match either color_1 or color_2
        binary_mask[np.all(rgb_image_array == color_1, axis=-1)] = 255
        binary_mask[np.all(rgb_image_array == color_2, axis=-1)] = 255
        binary_mask[np.all(rgb_image_array == color_3, axis=-1)] = 255
        binary_mask[np.all(rgb_image_array == color_4, axis=-1)] = 255

        # Convert the mask to have 4 channels (HxWx4)
        binary_mask_rgba = np.stack([binary_mask] * 4, axis=-1)

        # Convert back to an image
        binary_mask_image = Image.fromarray(binary_mask_rgba)

        # Save the binary mask image in the destination folder
        save_path = os.path.join(destination_folder, filename)
        binary_mask_image.save(save_path)

        print(f"Binary mask for {filename} created and saved successfully!")

print("All images processed successfully!")
```

```python
import os
import shutil

# Define folder paths
base_folder = 'data/f8-data-v2/2663250'
images_folder = os.path.join(base_folder, 'images_all')
masks_folder = os.path.join(base_folder, 'masks_rgb')
output_folder = os.path.join(base_folder, 'images_nohemo')

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Get list of image filenames without extension in masks folder (assuming .png files)
mask_files = {os.path.splitext(f)[0] for f in os.listdir(masks_folder) if f.endswith('.png')}

# Iterate through images folder and check for corresponding mask file
for image_file in os.listdir(images_folder):
    if image_file.endswith('.jpg'):
        image_name = os.path.splitext(image_file)[0]
        # If there is no corresponding mask file, copy the image to output folder
        if image_name not in mask_files:
            src_path = os.path.join(images_folder, image_file)
            dest_path = os.path.join(output_folder, image_file)
            shutil.copy(src_path, dest_path)
            print(f"Copied: {image_file}")

print("Script completed. Images without masks have been copied to 'images_nohemo'.")
```

```python
import os
from datetime import datetime

# Directory containing the files
directory = "/Users/jamesemilian/triage/hemobox/data/f8-data-v2/2663250_wcopy/images"

# Iterate through all files in the directory
for filename in os.listdir(directory):
    # Check if the file matches the pattern "image_YYYYMMDD_HHMMSS.png"
    if filename.startswith("image_") and filename.endswith(".png"):
        try:
            # Extract the datetime part from the filename
            parts = filename.split("_")
            date_part = parts[1]  # YYYYMMDD
            time_part = parts[2].replace(".png", "")  # HHMMSS

            # Combine and parse into a datetime object
            dt = datetime.strptime(date_part + time_part, "%Y%m%d%H%M%S")
            
            # Convert to Unix timestamp in nanoseconds
            unix_ns = int(dt.timestamp() * 1e9)  # Convert to nanoseconds
            
            # Create the new filename
            new_filename = f"{unix_ns}.png"
            
            # Rename the file
            old_path = os.path.join(directory, filename)
            new_path = os.path.join(directory, new_filename)
            os.rename(old_path, new_path)
            print(f"Renamed: {filename} -> {new_filename}")
        except Exception as e:
            print(f"Error processing {filename}: {e}")
```
